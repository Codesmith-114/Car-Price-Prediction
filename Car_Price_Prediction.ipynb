```python
# Basic cleaning - remove specific symbols
df = df.replace('?', np.nan)  # Replace ? with NaN
df = df.replace('[\$,]', '', regex=True)  # Remove $ and commas

# More comprehensive cleaning for all columns
def clean_column(col):
    if df[col].dtype == 'object':  # Only clean string columns
        df[col] = df[col].str.replace(r'[^\w\s]', '', regex=True)  # Remove all special characters
        df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert to numeric if possible

# Apply cleaning to all columns
for column in df.columns:
    clean_column(column)

# Check for any remaining missing values
print(df.isnull().sum())
```

```python
# Drop columns where all values are null
df = df.dropna(axis=1, how='all')

# Drop columns where more than 70% values are null
threshold = len(df) * 0.3  # keep columns with at least 30% non-null values
df = df.dropna(axis=1, thresh=threshold)

# To check the remaining null values in each column
null_counts = df.isnull().sum()
print("Remaining null values per column:")
print(null_counts[null_counts > 0])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = df.drop(columns=['price'])
y = df['price']
X_scaled = scaler.fit_transform(X)
X.head()
```

```python
pd.DataFrame(X_scaled, columns=X.columns, index=X.index).head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

```

```python
from sklearn.impute import SimpleImputer

# Impute missing values in X_train and X_test
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)
model = LinearRegression(n_jobs=-1)
model.fit(X_train_imputed, y_train)
y_pred = model.predict(X_test_imputed)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
r2_score(y_test, y_pred)
```

```python
sns.histplot(x=df['price'], kde=True, fill=True, label="Price")
sns.histplot(x=y_pred, kde=True, fill=True, label="Predicted Price", color="orange", alpha=0.5)
plt.legend()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
categorical_features = ['make', 'fuel-type', 'aspiration', 'num-of-doors', 
                        'body-style', 'drive-wheels', 'engine-type', 'fuel-system']

# Define the numerical features that will be standardized
numerical_features = ['wheel-base', 'length', 'width', 'height', 'curb-weight', 
                      'engine-size', 'bore', 'stroke', 'compression-ratio', 
                      'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg']
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)])

# Create a pipeline that first preprocesses the data and then can be used to fit a model
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
svr_model = SVR(kernel='rbf')

# Train the SVR model on the preprocessed training data
svr_model.fit(X_train_imputed, y_train)

# Predict the target values for the test set using the trained SVR model
y_pred_svr = svr_model.predict(X_test_imputed)
# Calculate the Mean Squared Error (MSE) and R-squared value for SVR
mse_svr = mean_squared_error(y_test, y_pred_svr)
r2_svr = r2_score(y_test, y_pred_svr)

# Print the evaluation metrics
print(f'SVR Mean Squared Error: {mse_svr}')
print(f'SVR R-Squared: {r2_svr}')
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the Random Forest model on the preprocessed training data
rf_model.fit(X_train_imputed, y_train)

# Predict the target values for the test set using the trained Random Forest model
y_pred_rf = rf_model.predict(X_test_imputed)
# Calculate the Mean Squared Error (MSE) and R-squared value for Random Forest
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

# Print the evaluation metrics
print(f'Random Forest Mean Squared Error: {mse_rf}')
print(f'Random Forest R-Squared: {r2_rf}')
```
